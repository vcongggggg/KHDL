```python
import pandas as pd

# 1. Đọc file CSV hiện tại của bạn
file_path = 'phone_row_add_brand.csv'
df = pd.read_csv(file_path)

# 2. Lọc bỏ các dòng bị khuyết (NaN) ở cột 'price'
# subset=['price'] giúp pandas chỉ kiểm tra lỗi thiếu dữ liệu ở đúng cột giá bán
df_cleaned = df.dropna(subset=['price'])

# 3. Xuất ra một file CSV mới đã được làm sạch hoàn toàn
output_file = 'phone_final_cleaned1.csv'
df_cleaned.to_csv(output_file, index=False, encoding='utf-8-sig')

# In ra thông báo kiểm tra
print(f"✅ Đã dọn dẹp xong!")
print(f"Số dòng ban đầu: {len(df)}")
print(f"Số dòng sau khi lọc: {len(df_cleaned)}")
print(f"File mới đã được lưu tại: {output_file}")
```
